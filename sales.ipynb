```python
import pandas as pd
```

```python
# uploaing data file to a pandas dataframe
data=pd.read_csv('transaction.csv',sep=';')
data.head()
```

```python
# Number of rows an colums in the dataframe respectively
data.shape
```

```python
# General info about columns
data.info()
```

```python
# Fetching maximum value in each column
a=data.max()
a
```

```python
# Found maximum value of year to be 2028
data.Year.unique()
```

```python
# Replacing maximum value of year 2028 with 2020
data['Year']=data['Year'].replace(2028,2020)
```

```python
# Now values are OK.
data.Year.unique()
```

```python
# Calculating New Columns--'Total_Cost_Price' , 'Total_Selling_Prce', 'Profit' , 'Markup', for clear insights of sales.
data['Total_Cost_Price']=data['NumberOfItemsPurchased']*data['CostPerItem']
data['Total_Selling_Price']=data['NumberOfItemsPurchased']*data['SellingPricePerItem']
data['Profit']=data['Total_Selling_Price']-data['Total_Cost_Price']
data['Markup']=data['Profit']/data['Total_Cost_Price']
```

```python
data.head()
```

```python
# Combining year,month,day to date
day=data['Day'].astype(str)
year=data['Year'].astype(str)
data['Date']=day+'-'+data['Month']+'-'+year
```

```python
data.head()
```

```python
# Splitting values in 'ClientKeywords'  into multiple columns for clear understanding of customer. 
split=data['ClientKeywords'].str.split(',',expand=True)
```

```python
split
```

```python
# Assigning the values now.
data['Client_Age']=split[0]
data['Client_Type']=split[1]
data['Length_Of_Contract']=split[2]
```

```python
data.head()
```

```python
# Removing '['  and']' from new columns 'Client_Age' and 'Length_Of Contract'
data['Client_Age']=data['Client_Age'].str.replace('[','')
data['Length_Of_Contract']=data['Length_Of_Contract'].str.replace(']','')
```

```python
data.head()
```

```python
# Changing 'Itemescription' values to lower case
data['ItemDescription']=data['ItemDescription'].str.lower()
```

```python
# Merging one data set which has seasons according to month
seasons=pd.read_csv('value_inc_seasons.csv',sep=';')  # loading to dataframe

seasons.head()
```

```python
# Merging
merge_df=pd.merge(data,seasons,how='left',on='Month')
```

```python
merge_df.head(100)
```

```python
# dropping columns of 'Day' ,'Year' , 'Month' from dataframe because they were merged together as 'Date' in previous step.
merge_df=merge_df.drop('Year',axis=1)
merge_df=merge_df.drop('Month',axis=1)
merge_df=merge_df.drop('Day',axis=1)
merge_df=merge_df.drop('ClientKeywords',axis=1)
```

```python
# exporting into a csv
merge_df.to_csv('mycleaned_data.csv',index=False)
```
